```python
athletes = athletes[athletes.Season == 'Summer']
```

```python
def team_summary(data):
    """ Summarizies the columns when will be grouped by country and year.
    """
    new_df = pd.Series({
        'team': data.iloc[0,:]['NOC'],
        'country': data.iloc[0,:]['Team'],
        'year': data.iloc[0,:]['Year'],
        'events': len(data['Event'].unique()),
        'athletes': len(data['Name']), # unique or not unique here ? 
        'age': data['Age'].mean(skipna=True),
        'height': data['Height'].mean(skipna=True),
        'weight': data['Weight'].mean(skipna=True),
        'medals': pd.notnull(data['Medal']).sum(),   
    })
    return new_df
    
```

```python
teams = athletes.groupby(["NOC", "Year"]).apply(team_summary)
teams
```

I now want to reset the indeces to a single list from 0 to len(team).

```python
teams = teams.reset_index(drop=True)
```

```python
teams.head()
```

Now I want to keep the data from last olympics, second last and third last.

```python
tmp = teams[teams['team']=='USA'].sort_values("year", ascending=True)
tmp.tail()
```

```python
teams = teams.dropna()
teams
```

```python
def previous_medals(data):
    """ Get number of medals from last year and an overage of the previous 3 editions
    """
    data = data.sort_values("year", ascending=True)
    # Ensure 'medals' is numeric to avoid aggregation errors
    data["medals"] = pd.to_numeric(data["medals"], errors="coerce")  # Convert, setting non-numeric to NaN

    data["prev_medals"] = data["medals"].shift(1)
    data["prev_3_medals"] = data["medals"].rolling(3, closed="left", min_periods=1).mean()
    
    return data
```

```python
teams = teams.groupby(['team']).apply(previous_medals)
```

```python
teams.head()
```

```python
teams = teams.reset_index(drop=True)
teams
```

```python
teams = teams[teams["year"] > 1960]
teams[teams.team == 'USA']
```

```python
teams = teams.round(1)
```
